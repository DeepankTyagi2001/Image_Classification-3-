#preprocessing

```python
import os
import matplotlib.pyplot as plt
import numpy as np
from skimage.io import imread
from skimage.transform import resize
from tqdm import tqdm
```

```python
target=[]
images=[]
flatten_images=[]

DATA_DIR="/content/images"

categories=["pretty sunflower","rugby ball","icecream cone"]

for cat in categories:
  class_num=categories.index(cat)
  path=os.path.join(DATA_DIR,cat)
  for img in tqdm(os.listdir(path)):
    image_array=imread(os.path.join(path,img))
    # print(image_array)
    # plt.imshow(image_array)
    # break
    resized_image=resize(image_array,(150,150,3))
    flatten_images.append(resized_image.flatten())
    images.append(resized_image)
    target.append(class_num)
```

#split data into training and testing

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(flatten_images,target,test_size=0.3,random_state=121)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn import svm
param_grid=[
    {
        'C':[1,10,100,1000],'kernel':['linear']
    },
     {
        'C':[1,10,100,1000],'gamma':[0.001,0.0001],'kernel':['linear']
    }
]
svc=svm.SVC(probability=True)
clf=GridSearchCV(svc,param_grid)
tqdm(clf.fit(x_train,y_train))
```

```python
y_predict=clf.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
accuracy_score(y_predict,y_test)
```

```python
confusion_matrix(y_predict,y_test)
```
